## 1.深度学习、机器学习、人工智能与神经网络
人工智能及相关领域的发展历程：
- 预测未来，需要在某个领域深入的“只适合经验”
- 经验和知识是无法在一夜之间获得
- 依靠概率计算、模式拟合（pattern fit）等方法,数学家们创造了许多数学模型（线性回归、贝叶斯模型等），这些模型能够根据历史数据学习出某种规律，并依赖这些规律实现预测
- 将能够根据历史数据实现预测或得出某种结果的计算步骤或计算方法称之为算法
- 计算机科学家将计算机与数学结合，实现算法的规模化，提升”学习“和”预测“的效率
- 哲学家与心理学家将大脑的思考机制复现，实现人工神经网络（ANN，1943年），来制造和人相似的智慧
- 1950年，提出“图灵测试”
- 1956年，提出人“人工智能”和“机器学习”概念
- 神经网络因为预测效果差、数据需求大、计算时间长、**根本无法达到人类智力水平**，遭遇寒冬
- 到新世纪，神经网络算法的大规模应用的条件逐渐成熟
  - 神经网络算法有了长足的进步
  - 全球数据量激增
  - 芯片、云技术迅猛发展，算力提升
- 2016年，提出“深度学习”概念

一般看来，**人工智能包含机器学习，机器学习包含深度学习**。

## 3.机器学习中的基本概念
深度学习中的概念与传统机器学习中有所区别。
（1）样本、特征与标签
- 样本：在高维张量的操作中，一般不再区分”行列“，而是认为每个索引对应的对象就是一个样本
- 特征：张量包含的内容就是特征，一般称特征所在的张量叫做特征张量
- 标签：一般与数据集市分开的

（2）分类与回归
标签是机器学习中非常重要的一个概念，不同的标签指向了不同的问题，应用最广泛的问题主要是两类：
- 分类：标签有限且互斥，标签是类别、表现为离散型变量（Categorical）
- 回归：模型输出是具体数字，标签是浮点数，表现为连续型变量（Continuous）

（3）有监督与无监督
- 有监督学习：有标签的任务，从已知的历史数据中进行学习，然后去预测渴望了解的东西，如KNN、决策树、支持向量机、线性回归、逻辑回归和大部分神经网络等
- 无监督学习：无标签的任务，一般作为作为辅助算法增强有监督算法的学习效果，包括聚类分析、协同过滤和变分自编码器等
- 半监督学习
- 强化学习

（4）模型的评估标准
- **模型预测效果**：模型进行判断/预测的效果一定是追求的核心目标，对于不同类型的算法，有不同的模型评估指标，我们依据这些评估指标来衡量模型的判断/预测效果
- **运算速度**：能够同时处理大量数据、可以在超短时间内极速学习、可以实时进行预测是机器学习的重要优势。如果算法的运算速度太慢，也不利于调优和实验，同时可能需要占用更多的计算和储存资源，带来更高的成本。在模型效果不错的情况下保障运算速度较快，是机器学习中重要的一环
- 可解释性：需要向人们解释算法的预测结果，否则相关人员河南接受。深度学习对模型的可解释性要求较低
- 服务于业务：只有服务于业务，或服务于推动人类认知的研究，算法才会具有商业价值

## 4.深入认识PyTorch框架
PyTorch框架的优势：
- 天生支持巨量数据和巨大神经网络的高速运算
- 灵活性高，足以释放神经网络的潜力，并且在保留灵活性的同时，又有Python语法简单易学的优势
- 支持研究环境与生产环境无缝切换，调试成本很低

PyTorch的基本架构：
- 原生Torch库——用于构建灵活神经网络的模块
  - 运行基础
    - tensor
    - autograd
  - 数据预处理utils
    - 数据导入与处理data、datasets
    - 可视化tensorboard
    - 预训练模型model_zoo
  - 神经网络基本元素nn
    - 各种层Module
    - 损失函数与激活函数functional
  - 优化算法optim
  - 运算性能
    - 分布式训练torchelastic
    - GPU训练cuda
  - 生产环境部署JIT
- 成熟AI领域用以辅助具体行业应用的模块
  - 计算机视觉torchvision
    - 常用数据集datasets
    - 常用模型models
    - 图像数据预处理transform
  - 自然语言处理torchtext
    - 数据预处理data
    - 常用数据集datasets
  - 语音处理torchaudio
    - 常用数据集datasets
    - 声音数据预处理transform
    - 常用模型models
    - 常用函数functional
